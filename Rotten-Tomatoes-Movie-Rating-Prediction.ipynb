## As Y or audience rating is continous so we will apply regression algorithms

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split


# Assuming you have your dataframe as `df_feature` and already split into X and Y
# X_train, X_test, y_train, y_test = train_test_split(df_feature.drop('target_column', axis=1), df_feature['target_column'], test_size=0.2, random_state=42)

# Identify categorical columns and numerical columns
categorical_columns = ['movie_title', 'genre', 'directors', 'writers', 'studio_name']
numerical_columns = ['runtime_in_minutes', 'tomatometer_rating', 'tomatometer_count', 'G', 'NC17',
       'NR', 'PG', 'PG-13', 'PG-13)', 'R', 'R)', 'tomatometer_status']

# Convert any non-numeric columns to string to handle mixed types
for col in categorical_columns:
    X_train[col] = X_train[col].astype(str)
    X_test[col] = X_test[col].astype(str)

# # For numerical columns, we can use mean or median to impute missing values
# imputer_num = SimpleImputer(strategy='mean')

# # Apply imputer to numerical columns
# X_train[numerical_columns] = imputer_num.fit_transform(X_train[numerical_columns])
# X_test[numerical_columns] = imputer_num.transform(X_test[numerical_columns])

# Check for missing values in numerical columns before applying imputer
missing_values = X_train[numerical_columns].isnull().sum()
print("Missing values in numerical columns:\n", missing_values)

# Only apply imputer if there are missing values
if missing_values.sum() > 0:
    imputer_num = SimpleImputer(strategy='mean')
    X_train[numerical_columns] = imputer_num.fit_transform(X_train[numerical_columns])
    X_test[numerical_columns] = imputer_num.transform(X_test[numerical_columns])
else:
    print("No missing values in numerical columns. Proceeding without imputation.")

    
    
# Proceed with the rest of the pipeline

# Create a column transformer for categorical and numerical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns),
        ('num', StandardScaler(), numerical_columns),
    ],
    remainder='passthrough'  # Keep any other columns as they are
)

# Create a pipeline with preprocessing and regression model
# pipeline = Pipeline(steps=[
#     ('preprocessor', preprocessor),
#     ('regressor', LinearRegression())
# ])

# Create a pipeline with preprocessing and linear regression model
pipeline_linear = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Create a pipeline with preprocessing and ridge regression model
pipeline_ridge = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=1.0))  # You can adjust alpha
])

# Create a pipeline with preprocessing and lasso regression model
pipeline_lasso = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=1.0))  # You can adjust alpha
])

# Create a pipeline with preprocessing and elastic net regression model
pipeline_elasticnet = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', ElasticNet(alpha=1.0, l1_ratio=0.5))  # You can adjust alpha and l1_ratio
])

# Create a pipeline with preprocessing and decision tree regression model
pipeline_tree = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeRegressor())
])

# Create a pipeline with preprocessing and random forest regression model
pipeline_forest = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100))  # You can adjust n_estimators
])

# Create a pipeline with preprocessing and gradient boosting regression model
pipeline_gb = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor())
])

# Create a pipeline with preprocessing and support vector regression model
pipeline_svr = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', SVR(kernel='rbf'))  # You can adjust the kernel
])


```

```python
# List of models to evaluate
pipelines = {
    'Linear Regression': pipeline_linear,
    'Ridge Regression': pipeline_ridge,
    'Lasso Regression': pipeline_lasso,
    'ElasticNet Regression': pipeline_elasticnet,
    'Decision Tree': pipeline_tree,
    'Random Forest': pipeline_forest,
    'Gradient Boosting': pipeline_gb,
    'Support Vector Regression': pipeline_svr
}

for name, pipeline in pipelines.items():
    pipeline.fit(X_train, Y_train)
    y_pred = pipeline.predict(X_test)
    
    mse = mean_squared_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)
    
    print(f'{name} - Mean Squared Error: {mse}, R^2 Score: {r2}')
```
